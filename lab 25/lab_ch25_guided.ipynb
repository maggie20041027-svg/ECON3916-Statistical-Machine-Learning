## Part 6: Reflection (10 min)

Answer each question in **3-5 sentences**. Connect your answers to concepts from Chapter 25.

### Q1: What worked well with AI-assisted development?

*Think about: speed of scaffolding, code quality, design suggestions, handling of repetitive tasks.*

**Your answer:** The scaffolding speed was the biggest win. Claude Code generated a complete Next.js site,  hero, navbar, projects grid, skills section, contact, and a working dark/light toggle, in one prompt and about eight minutes of build time, which would have taken me days to write from scratch given I don't know React or Tailwind. It also handled repetitive pattern-heavy tasks extremely well, like turning a list of project descriptions into uniform card components with consistent styling. Design decisions like typography, spacing, and color palette were reasonable out of the box, so I mostly spent my time on content rather than construction. This matches the Chapter 25 point that AI is most useful when it handles the boilerplate and lets the human focus on judgment and substance.



### Q2: What did you have to fix manually?

*Think about: layout bugs, incorrect assumptions, styling mismatches, missing features, hallucinated APIs.*

**Your answer:** The biggest "manual" fixes were actually environment problems the AI couldn't have predicted: I didn't have Node.js or Homebrew installed, npm install -g failed on a permissions error requiring sudo, GitHub rejected password authentication, and gh wasn't installed either, each of these cost me more time than the actual coding did. On the content side, Claude Code initially created ten placeholder project cards when I only had a handful of real labs to showcase, so I had to feed it the real GitHub repo contents for it to write accurate descriptions instead of hallucinated ones. The AI also made confident but wrong assumptions about what my labs were "about" until I gave it concrete numbers and findings — it's good at polishing, but it needs the truth to polish in the first place.



### Q3: How does AI-assisted development compare to writing code entirely yourself?

*Think about: the HADI loop from Chapter 25 (Hypothesis, Action, Data, Insight). Did you follow this pattern? Where did you need domain knowledge that the AI lacked?*

**Your answer:** Writing it myself would have been a linear grind through documentation; AI-assisted development felt more like the HADI loop from Chapter 25 — I formed a hypothesis ("the hero should feel editorial, not startup-y"), took an action (prompting Claude Code), saw the data (the rendered site in my browser), and derived insight (what to change next). The loop cycled in minutes rather than hours, which meant I could iterate on design and content decisions that I'd normally never bother with. Where the AI fell short was anywhere domain knowledge was load-bearing: it couldn't decide which of my labs were strongest, couldn't write project descriptions that accurately conveyed what I learned in Lab 24, and couldn't tell whether a particular phrase sounded like an economics student versus a generic CS grad. Those judgments — what to showcase, how to frame it, what to leave out, still had to come from me.



### Q4: Prompt engineering lessons

*What makes a good prompt for Claude Code? What makes a bad one? Give one specific example of a prompt you improved.*

**Your answer:** Good prompts are specific and grounded in reality; bad prompts are vague or assume the model knows context it doesn't have. An early bad prompt of mine was asking Claude Code to "add my real projects" — it invented plausible-sounding but fictional descriptions because I hadn't given it anything concrete to work with. I improved it to: "Fetch the contents of my GitHub repo at the specific websit, find the folders for Labs 20–24, and write 1–2 sentence descriptions for each based on the README and notebook contents, then replace the placeholder projects in Projects.tsx." That version worked on the first try because the AI had a clear data source, a clear scope, and a clear output format. The lesson: treat prompts like specs for a junior engineer — if a human would need clarification, so will the model.



### Q5: If you were hiring a junior data scientist, would their portfolio site influence your evaluation? Why or why not?

*Connect to the "signaling" concept from labor economics — what does a portfolio signal that a resume alone does not?*

**Your answer:** Yes, significantly, and the reason connects directly to the signaling concept from labor economics. A résumé lists coursework and GPA, but those are noisy signals easily gamed, while a deployed portfolio is a costly-to-fake signal: it simultaneously demonstrates technical execution (the site actually works), communication skill (the project descriptions are coherent), and basic professionalism (the thing deploys on its own custom URL). For a junior data scientist in particular, the portfolio separates candidates who have done analyses from those who have only read about them — because anyone can claim "I used Double Machine Learning in a class project," but a linked, running repo with real outputs is much harder to bluff. If two applicants had identical résumés and one had a live portfolio showing working DML code, I would interview that person first.
